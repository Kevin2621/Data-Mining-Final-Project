# Cleaning all data from "heart-disease-mortality-by-location(2012-2014).csv"

```python
data_2012_2014 = pd.read_csv("heart-disease-mortality-by-location(2012-2014).csv")

data_2012_2014 = data_2012_2014.rename(columns={'location_1': 'LatLongPoint', 'locationid': 'FIPS', 'locationdesc': 'County/Region Name', 'data_value': 'Cases Per 100,000 Population', 'stratification1': 'Gender', 'locationabbr': 'State/Territory Abbriviation', 'stratification2': 'Race/Ethnicity'})

# unique = data['topic'].unique()

# create lat and long columns from the combinded lat long points
data_2012_2014['longitude'] = data_2012_2014['LatLongPoint'].str.extract(r'POINT \((-?\d+\.\d+)')[0].astype(float)
data_2012_2014['latitude'] = data_2012_2014['LatLongPoint'].str.extract(r'POINT \(-?\d+\.\d+ (\-?\d+\.\d+)\)')[0].astype(float)

# drop unneccessary columns
data_2012_2014 = data_2012_2014.drop(columns=['data_value_unit', 'topic', 'stratificationcategory1', 'data_value_footnote', 'data_value_footnote_symbol', 'location_1_zip', 'location_1_city', 'location_1_state', 'location_1_address'])

data_2012_2014


```

```python

# Filter the data where race/ethnicity is "Overall"
overall_data_2012_2014 = data_2012_2014[(data_2012_2014['Race/Ethnicity'] == 'Overall') & (data_2012_2014['Gender'] == 'Overall')]

# Plot the distribution
plt.figure(figsize=(10, 6))
plt.hist(overall_data_2012_2014['Cases Per 100,000 Population'], bins=100, edgecolor='black')
plt.title('Distribution of Cardiovascular Disease Cases Per 100,000 Population')
plt.xlabel('Cases Per 100,000 Population')
plt.ylabel('Frequency')

# plt.xlim(0, 1500)
plt.show()
```

```python
# Apply log transformation and plot histograms and box plots for each DataFrame
for period, df in dataframes.items():
    # Make a copy of the DataFrame to avoid SettingWithCopyWarning
    df = df.copy()
    
    # Apply log transformation
    df['Log_Cases'] = np.log1p(df['Cases Per 100,000 Population'])  # log1p to handle zeros

    # Histogram of log-transformed data
    plt.figure(figsize=(10, 6))
    plt.hist(df['Log_Cases'], bins=50, edgecolor='black')
    plt.title(f"Log-Transformed Distribution of Heart Disease Cases per 100,000 Population ({period})")
    plt.xlabel("Log(Cases per 100,000 Population)")
    plt.ylabel("Frequency")
    plt.show()

    # Box Plot of log-transformed data
    plt.figure(figsize=(8, 6))
    plt.boxplot(df['Log_Cases'], vert=False)
    plt.title(f"Box Plot of Log-Transformed Heart Disease Cases per 100,000 Population ({period})")
    plt.xlabel("Log(Cases per 100,000 Population)")
    plt.show()
```

```python
def create_map(data, title, threshold_scale):
    m = folium.Map(location=[37.0902, -95.7129], zoom_start=5)

    geo_json_url = "https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json"

    folium.Choropleth(
        geo_data=geo_json_url,
        name="choropleth",
        data=data,
        columns=["FIPS", "Cases Per 100,000 Population"],
        key_on="feature.id",
        fill_color="YlOrRd",
        fill_opacity=0.7,
        line_opacity=0.2,
        legend_name=title,
        threshold_scale=threshold_scale
    ).add_to(m)

    display(m)
    

# Calculate min and max for each dataset
min_values = [
    overall_data_2012_2014['Cases Per 100,000 Population'].min(),
    overall_data_2015_2017['Cases Per 100,000 Population'].min(),
    overall_data_2017_2019['Cases Per 100,000 Population'].min(),
    overall_data_2019_2021['Cases Per 100,000 Population'].min()
]

max_values = [
    overall_data_2012_2014['Cases Per 100,000 Population'].max(),
    overall_data_2015_2017['Cases Per 100,000 Population'].max(),
    overall_data_2017_2019['Cases Per 100,000 Population'].max(),
    overall_data_2019_2021['Cases Per 100,000 Population'].max()
]

# Calculate the global min and max
global_min = min(min_values) * 0.99
global_max = max(max_values) * 1.01

# Define threshold scale with 5 evenly spaced values from global_min to global_max
threshold_scale = list(np.linspace(global_min, global_max, num=10))

# Display each map in a separate cell
create_map(overall_data_2012_2014, "Cardiovascular Diseases Cases Per 100,000 Population (2012-2014)", threshold_scale)
create_map(overall_data_2015_2017, "Cardiovascular Diseases Cases Per 100,000 Population (2015-2017)", threshold_scale)
create_map(overall_data_2017_2019, "Cardiovascular Diseases Cases Per 100,000 Population (2017-2019)", threshold_scale)
create_map(overall_data_2019_2021, "Cardiovascular Diseases Cases Per 100,000 Population (2019-2021)", threshold_scale)
```

```python
# Define an upper limit to cap population values for better contrast
upper_limit = 10000000  # Adjust based on data range and visualization needs

# Define a function to create histograms and box plots in groups of 4 years with log scaling and capping
def create_population_plots(data):
    for group in range(1, 4):  # There are three groups: 1-4, 5-8, 9-12
        fig, axes = plt.subplots(2, 4, figsize=(20, 10))  # 2x4 grid (histogram and box plot for each year)
        
        # Loop over the years in the current group
        for i, year in enumerate(range((group - 1) * 4 + 1, group * 4 + 1)):
            # Filter and cap data for the current year
            year_data = data[data['YEAR'] == year]['TOT_POP']
            capped_data = year_data.apply(lambda x: min(x, upper_limit))  # Cap high population values
            
            # Histogram (top row) - Distribution of population across counties for this year
            axes[0, i].hist(capped_data, bins=50, edgecolor='black', log=True)  # Use log scale
            axes[0, i].set_title(f'Population Distribution (Year {2009 + year})')
            axes[0, i].set_xlabel('Total Population per County (Capped)')
            axes[0, i].set_ylabel('Frequency')
            
            # Turn off scientific notation on the x-axis
            axes[0, i].ticklabel_format(style='plain', axis='x')
            
            # Box plot (bottom row) - Population spread across counties for this year
            axes[1, i].boxplot(capped_data.dropna(), vert=False)
            axes[1, i].set_xscale('log')  # Use log scale for box plot
            axes[1, i].set_title(f'Box Plot of Population (Year {2009 + year})')
            axes[1, i].set_xlabel('Total Population per County (Log Scale, Capped)')
            
        plt.tight_layout()
        plt.show()

# Filter data to include only rows where AGEGRP is 0 (total population per county)
pop_tot_demographics_2010_2019 = pop_demographics_2010_2019[['YEAR', 'STATE', 'COUNTY', 'TOT_POP', 'AGEGRP']]
pop_tot_demographics_2010_2019 = pop_tot_demographics_2010_2019[pop_tot_demographics_2010_2019['AGEGRP'] == 0]

# Create a FIPS column by combining STATE and COUNTY codes, padded to ensure it's 5 digits
pop_tot_demographics_2010_2019['FIPS'] = pop_tot_demographics_2010_2019['STATE'].astype(str).str.zfill(2) + pop_tot_demographics_2010_2019['COUNTY'].astype(str).str.zfill(3)

# Call the function to create and display the plots
create_population_plots(pop_tot_demographics_2010_2019)
```

```python
# Filter data to include only rows where AGEGRP is 0 (total population)
data = pop_tot_demographics_2010_2019[pop_tot_demographics_2010_2019['AGEGRP'] == 0]

# Create a FIPS column by combining STATE and COUNTY codes, padded to ensure it's 5 digits
data['FIPS'] = data['STATE'].astype(str).str.zfill(2) + data['COUNTY'].astype(str).str.zfill(3)

# Define an upper limit for the highest category (e.g., 500,000) for better contrast
upper_limit = 500000  # Adjust as needed based on your data

# Define the years to display (first and last)
years_to_display = [1, 12]  # 1 represents 2010, 12 represents 2021

# Loop through the selected years to create the heat maps with capped threshold scale
for year in years_to_display:
    # Filter data for the specific year
    year_data = data[data['YEAR'] == year].copy()  # Use .copy() to avoid SettingWithCopyWarning
    
    # Cap the population values at the upper limit to avoid high values washing out the lower values
    year_data.loc[:, 'Capped_POP'] = year_data['TOT_POP'].apply(lambda x: min(x, upper_limit))

    # Define a custom threshold scale with more focus on lower population ranges and a cap for high values
    threshold_scale = [0, 10000, 50000, 100000, 200000, upper_limit]  # Customize these cutoffs as needed

    # Load the GeoJSON for U.S. counties
    geo_json_url = "https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json"
    
    # Initialize the map centered on the United States
    m = folium.Map(location=[37.0902, -95.7129], zoom_start=5)
    
    # Add a choropleth layer for the total population in each county
    folium.Choropleth(
        geo_data=geo_json_url,
        name="choropleth",
        data=year_data,
        columns=["FIPS", "Capped_POP"],
        key_on="feature.id",
        fill_color="YlOrRd",
        fill_opacity=0.7,
        line_opacity=0.2,
        threshold_scale=threshold_scale,
        legend_name=f"Total Population by County (Year {2009 + year}) - Capped at {upper_limit}"
    ).add_to(m)
    
    # Display the map for each selected year
    display(m)
```
